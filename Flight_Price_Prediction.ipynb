```python
x=df.drop(['price'],axis=1, inplace = False)
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,1))
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
x_train=pd.DataFrame(x_train)
x_test=pd.DataFrame(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn import linear_model
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
import xgboost as xgb
```

```python
from sklearn import metrics
```

```python
def rmse(targets, predictions):
  return np.sqrt(metrics.mean_squared_error(targets,predictions))
```

```python
def r2(targets,predictions):
  return(metrics.r2_score(targets,predictions))
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr_preds = lr.predict(x_test)
```

```python
rmse(y_test,lr_preds)
```

```python
r2(y_test,lr_preds)
```

```python
ridge = Ridge()
ridge.fit(x_train, y_train)

tree = DecisionTreeRegressor()
tree.fit(x_train, y_train)

forest = RandomForestRegressor()
forest.fit(x_train, y_train)

xgb = xgb.XGBRegressor()
xgb.fit(x_train, y_train)
```

```python
print("RMSE for Linear Regression:- " , rmse(y_test,lr_preds))
print("R2 for Linear Regression :- ", r2(y_test,lr_preds))

ridge_preds = ridge.predict(x_test)
print("RMSE for Ridge:- " , rmse(y_test,ridge_preds))
print("R2 for Ridge :- ", r2(y_test,ridge_preds))

tree_preds = tree.predict(x_test)
print("RMSE for Decision Tree :- " , rmse(y_test,tree_preds))
print("R2 for Decision Tree :- ", r2(y_test,tree_preds))

forest_preds = forest.predict(x_test)
print("RMSE for Random Forest :- " , rmse(y_test,forest_preds))
print("R2 for Random Forest :- ", r2(y_test,forest_preds))

xgb_preds = xgb.predict(x_test)
print("RMSE for XGB :- " , rmse(y_test,xgb_preds))
print("R2 for XGB :- ", r2(y_test,xgb_preds))
```

```python
forest.score(x_test,y_test)
```

```python
out=pd.DataFrame({'Price_actual':y_test,'Price_pred':forest_preds})
result=df.merge(out,left_index=True,right_index=True)
```

```python
model = joblib.load("flight_price_model.pkl")
scaler = joblib.load("scaler.pkl")


le = LabelEncoder()


def preprocess_new_input(df):

    if 'Unnamed: 0' in df.columns:
        df.drop(['Unnamed: 0'], axis=1, inplace=True)


    for col in df.columns:
        if df[col].dtype == 'object':
            df[col] = le.fit_transform(df[col].astype(str))


    df_scaled = scaler.transform(df)

    return df_scaled


new_data = pd.DataFrame([{
    'airline': 'IndiGo',
    'flight': '6E-202',
    'source_city': 'Delhi',
    'departure_time': 'Evening',
    'stops': 'one',
    'arrival_time': 'Night',
    'destination_city': 'Cochin',
    'class': 'Economy',
    'duration': 2.8,
    'days_left': 15
}])


new_data_processed = preprocess_new_input(new_data)


predicted_price = model.predict(new_data_processed)

print("Predicted Flight Price: â‚¹", round(predicted_price[0], 2))
```
